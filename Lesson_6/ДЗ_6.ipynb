```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
df
```

```python
y = df.iloc[:,-1]
X = df.iloc[:,:56]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
def linear_regression(X, y):
    X = np.array(X).tolist() #приводим в формат списка
    y = np.array(y).tolist() #приводим в формат списка
    X_transpose = np.transpose(X) #транспонируем матрицу
    X_transpose_X_inv = np.linalg.inv(np.dot(X_transpose, X))# произведение транспонированной матрицы признаков X на саму матрицу признаков X
    # вычисляем обратную матрицу.
    X_transpose_y = np.dot(X_transpose, y) #умнажаем матрицы X транспонированную и у
    coefficients = np.dot(X_transpose_X_inv, X_transpose_y)
    return coefficients

# Обучение модели и получение коэффициентов
coefficients = linear_regression(X, y)

print("Коэффициенты модели:", coefficients)
```

```python
def predict(coefficients, X):
    predictions = []
    X = np.array(X).tolist()
    for i in range(len(X)):
        prediction = sum(coefficients[j] * X[i][j] for j in range(len(X[0])))
        predictions.append(prediction)
    return predictions

# Получение прогнозов
predictions = predict(coefficients, X_test)
```

```python
from sklearn.metrics import  mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
r2 = r2_score(y_test, predictions) 
r2
```
